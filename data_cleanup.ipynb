```python
# Importing modules
import pandas as pd
```

```python
# Read CSV
orig_data_df = pd.read_csv('Data\AviationData_Only_US.csv', engine='python')

# Display DF
orig_data_df
```

```python
# List the columns in the DF
orig_data_df.columns
```

```python
# Dropping the columns which are not necessary
orig_data_df.drop(['Registration.Number', 'Amateur.Built', 
                   'FAR.Description', 'Publication.Date', 'Schedule',
                   'Air.Carrier', 'Publication.Date'],
                   axis = 1, inplace = True)
```

```python
# Display DF
orig_data_df
```

```python
# Rename columns to replace . with _ in the column name
orig_data_df.rename(columns={'Event.Id' : 'Event_Id', 
                             'Investigation.Type' : 'Investigation_Type',
                             'Accident.Number' : 'Accident_Number',
                             'Event.Date' : 'Event_Date',
                             'Airport.Code' : 'Airport_Code',
                             'Airport.Name' : 'Airport_Name',
                             'Aircraft.Category' : 'Aircraft_Category',
                             'Injury.Severity' : 'Injury_Severity',
                             'Aircraft.Damage' : 'Aircraft_Damage',
                             'Number.of.Engines' : 'Number_of_Engines',
                             'Engine.Type' : 'Engine_Type',
                             'Purpose.of.Flight' : 'Purpose_of_Flight',
                             'Total.Fatal.Injuries' : 'Total_Fatal_Injuries',
                             'Total.Serious.Injuries' : 'Total_Serious_Injuries',
                             'Total.Minor.Injuries' : 'Total_Minor_Injuries',
                             'Total.Uninjured' : 'Total_Uninjured',
                             'Weather.Condition' : 'Weather_Condition',
                             'Broad.Phase.of.Flight' : 'Broad_Phase_of_Flight',
                             'Report.Status' : 'Report_Status'
                            }, inplace = True)

# Display DF
orig_data_df
```

```python
# Get the unique list of values from injury_severity column
orig_data_df['Injury_Severity'].unique()
```

```python
# As the Fatal value in Injury_Severity column has the number of fatal injury 
# count which we already have in another column.
# Replacing all these values with Fatal

orig_data_df['Injury_Severity'] = orig_data_df['Injury_Severity'].replace( ['Fatal(1)', 'Fatal(2)',
       'Fatal(3)', 'Fatal(4)', 'Fatal(5)', 'Fatal(9)', 'Fatal(7)',
       'Fatal(10)', 'Fatal(11)', 'Fatal(6)', 'Fatal(16)', 'Fatal(14)',
       'Fatal(50)', 'Fatal(8)', 'Fatal(49)', 'Fatal(20)', 'Fatal(13)',
       'Fatal(21)', 'Fatal(265)', 'Fatal(44)', 'Fatal(64)', 'Fatal(65)',
       'Fatal(92)', 'Fatal(18)', 'Fatal(19)', 'Fatal(88)', 'Fatal(228)',
       'Fatal(29)', 'Fatal(230)', 'Fatal(110)', 'Fatal(12)', 'Fatal(15)',
       'Fatal(68)', 'Fatal(132)', 'Fatal(37)', 'Fatal(27)', 'Fatal(23)',
       'Fatal(25)', 'Fatal(34)', 'Fatal(73)', 'Fatal(111)', 'Fatal(43)',
       'Fatal(28)', 'Fatal(156)', 'Fatal(82)', 'Fatal(17)', 'Fatal(31)',
       'Fatal(135)', 'Fatal(70)', 'Fatal(153)', 'Fatal(78)'], 'Fatal')
```

```python
# Verifying if all the values are replaced.
orig_data_df['Injury_Severity'].unique()
```

```python
# Copyting data to another df to get the perfect data set
perfect_data_df = orig_data_df.copy()

# Displat new DF
perfect_data_df
```

```python
# Check for missing values in new DF
perfect_data_df.count()
```

```python
# Drop all rows which includes NaN values
perfect_data_df = perfect_data_df.dropna()
```

```python
# Verifyng count after dropping all NaN
perfect_data_df.count()
```

```python
# Save clean file after cleanup
perfect_data_df.to_csv("Data\Perfect_AviationData.csv", index=False, header=True)
```

```python
# Now in Original DF replace empty string values with "Unknown" 
orig_data_df[['Investigation_Type', 'Location', 'State', 'Airport_Code',
                'Airport_Name', 'Aircraft_Damage', 'Aircraft_Category', 'Make',
                'Model', 'Engine_Type', 'Purpose_of_Flight', 'Weather_Condition',
                'Broad_Phase_of_Flight', 'Report_Status'
              ]] = orig_data_df[['Investigation_Type', 'Location', 'State', 'Airport_Code',
                'Airport_Name', 'Aircraft_Damage', 'Aircraft_Category', 'Make',
                'Model', 'Engine_Type', 'Purpose_of_Flight', 'Weather_Condition',
                'Broad_Phase_of_Flight', 'Report_Status'
              ]].fillna('Unknown')
```

```python
# Verify count
orig_data_df.count()
```

```python
# In Original DF replace empty numeric values with "0" except for columns for lat & long
orig_data_df[[ 'Injury_Severity', 'Number_of_Engines', 'Total_Fatal_Injuries', 
               'Total_Serious_Injuries', 'Total_Minor_Injuries', 'Total_Uninjured'
              ]] = orig_data_df[[ 'Injury_Severity', 'Number_of_Engines', 'Total_Fatal_Injuries', 
               'Total_Serious_Injuries', 'Total_Minor_Injuries', 'Total_Uninjured'
              ]].fillna(0) 
```

```python
# Verify count again
orig_data_df.count()
```

```python
# Drop all NaN values from DF now
orig_data_df = orig_data_df.dropna()
```

```python
# Re-verify count to confirm if count for all column values is same
orig_data_df.count()
```
